```python
# Verificación de instalación de librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

print("NumPy version:", np.__version__)
print(" Pandas version:", pd.__version__)
print("Matplotlib version:", plt.matplotlib.__version__)
print("Seaborn version:", sns.__version__)
print(" Scikit-learn version:", sklearn.__version__)
print("\n ¡Todo está instalado correctamente!")
```

```python
# Mini prueba de K-means con scikit-learn (solo para verificar)
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generamos datos de prueba
X, _ = make_blobs(n_samples=150, centers=3, random_state=42)

# Aplicamos K-means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

# Visualizamos
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c='red', marker='X', s=200, label='Centroides')
plt.title('Prueba de K-means - Si ves esto, ¡todo funciona!')
plt.legend()
plt.show()
```
